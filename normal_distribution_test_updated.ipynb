```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.formula.api import ols
from statsmodels.graphics.gofplots import qqplot

# Load Data
hardi_data = pd.read_csv("~/Downloads/NfL/currently_working_on/hardi_dti_data.csv")
dbsi_data = pd.read_csv("~/Downloads/NfL/currently_working_on/dbsi_data.csv")
orig_data = pd.read_csv("~/Downloads/NfL/currently_working_on/data_demo+nfl+vol+mwf.csv")
#Combine three datasets: HARDI, DBSI, Demographics+Volume+NfL+MWF data
DF = pd.merge(pd.merge(orig_data, dbsi_data, how='left', on='ID'), hardi_data, how='left', on='ID')
```

```python
for i in list(DF):
    if not (i in ['ID','Type','Subtype','Sex']):
        print(i)
        s_stat, s_p = stats.shapiro(DF[i].dropna())
        print('Shapiro:    Statistics=%.3f, p=%.3f' % (s_stat, s_p))
        d_stat, d_p = stats.normaltest(DF[i].dropna())
        print('D’Agostino’s K^2:    Statistics=%.3f, p=%.3f' % (d_stat, d_p))
        a_result = stats.anderson(DF[i].dropna())
        print('Anderson-Darling:    Statistic: %.3f' % a_result.statistic)
        for n in range(len(a_result.critical_values)):
            sl, cv = a_result.significance_level[n], a_result.critical_values[n]
            if a_result.statistic < a_result.critical_values[n]:
                print('%.3f: %.3f, data looks normal (fail to reject H0)' % (sl, cv))
            else:
                print('%.3f: %.3f, data does not look normal (reject H0)' % (sl, cv))
        qqplot(DF[i].dropna(), line='s', fit=True)
        plt.show()
```

```python
qqplot(np.log2(DF['NfL'].dropna().values), line='s', fit=True)
plt.show()
s_stat, s_p = stats.shapiro(np.log2(DF['NfL'].dropna().values))
print('Shapiro:    Statistics=%.3f, p=%.3f' % (s_stat, s_p))
d_stat, d_p = stats.normaltest(np.log2(DF['NfL'].dropna().values))
print('D’Agostino’s K^2:    Statistics=%.3f, p=%.3f' % (d_stat, d_p))
a_result = stats.anderson(np.log2(DF['NfL'].dropna().values))
print('Anderson-Darling:    Statistic: %.3f' % a_result.statistic)
for n in range(len(a_result.critical_values)):
    sl, cv = a_result.significance_level[n], a_result.critical_values[n]
    if a_result.statistic < a_result.critical_values[n]:
        print('%.3f: %.3f, data looks normal (fail to reject H0)' % (sl, cv))
    else:
        print('%.3f: %.3f, data does not look normal (reject H0)' % (sl, cv))
```
